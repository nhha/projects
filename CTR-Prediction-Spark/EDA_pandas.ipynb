```python
app_id_click = data[['app_id','click']].groupby('app_id').agg(['count','sum'])
app_id_click.sort_values(('click','count'), ascending=False).head(20)
```

```python
site_id_click = data[['site_id','click']].groupby('site_id').agg(['count','sum'])
site_id_click.sort_values(('click','count'), ascending=False).head(20)
```

```python
device_id_click = data[['device_id','click']].groupby('device_id').agg(['count','sum'])
device_id_click.sort_values(('click','count'), ascending=False).head(20)
```
